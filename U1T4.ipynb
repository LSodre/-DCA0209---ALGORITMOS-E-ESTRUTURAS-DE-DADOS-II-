```python
import csv

class Inventory:
    def __init__(self, csv_filename):
        with open(csv_filename) as file:
            csv_reader = csv.reader(file)
            csv_data = list(csv_reader)
       
        self.header = csv_data[0]
        self.rows = csv_data[1:]

        # Converte o preço de cada laptop para um valor inteiro
        for row in self.rows:
            row[-1] = int(row[-1])

        # Cria um dicionário para mapear IDs de laptops às linhas correspondentes de dados
        self.id_to_row = {}
        for row in self.rows:
            self.id_to_row[row[0]] = row 

        # Cria um conjunto para armazenar os preços únicos de laptops
        self.unique_prices = set()
        for row in self.rows:
            self.unique_prices.add(row[-1])

    # Obtém informações de um laptop com base no ID
    def get_laptop_from_id(self, laptop_id):
        for row in self.rows:
            if row[0] == laptop_id:
                return row
        return None   

    # Método otimizado para obter informações de um laptop com base no ID
    def get_laptop_from_id_fast(self, laptop_id):  
        return self.id_to_row.get(laptop_id)

    # Verifica se há promoções de laptops com preços específicos
    def check_promotion_dollars(self, dollars):    
        for row in self.rows:                   
            if row[-1] == dollars:
                return True
        for row1 in self.rows:                  
            for row2 in self.rows:
                if row1[-1] + row2[-1] == dollars:
                    return True
        return False
    
    # Método otimizado para verificar se há promoções de laptops com preços específicos
    def check_promotion_dollars_fast(self, dollars):
        if dollars in self.unique_prices:
            return True
        for price in self.unique_prices:
            if dollars - price in self.unique_prices:
                return True
        return False
    
    # Encontra laptops dentro de uma faixa de preço específica
    def find_laptops_in_price_range(self, min_price, max_price): 
        laptops_in_range = []
        for row in self.rows:
            price = row[-1]
            if min_price <= price <= max_price:
                laptops_in_range.append(row)
        return laptops_in_range
    
    # Encontra o laptop mais barato com especificações mínimas de RAM e armazenamento
    def find_cheapest_laptop_with_specifications(self, min_ram_gb, min_storage_gb):
        matching_laptops = []
        for row in self.rows:
            ram_str = row[7]
            storage_str = row[8]
            ram_gb = int(ram_str.split('GB')[0])
            storage_capacity_gb = 0
            
            if 'TB' in storage_str:
                storage_capacity_str = storage_str.split('TB')[0]
                if storage_capacity_str.isdigit():
                    storage_capacity_gb = int(storage_capacity_str) * 1000
            elif 'GB' in storage_str:
                storage_capacity_str = storage_str.split('GB')[0]
                if storage_capacity_str.isdigit():
                    storage_capacity_gb = int(storage_capacity_str)

            if ram_gb >= min_ram_gb and storage_capacity_gb >= min_storage_gb:
                matching_laptops.append(row)

        if matching_laptops:
            cheapest_laptop = min(matching_laptops, key=lambda x: x[-1])
            return cheapest_laptop
        else:
            return None


inventario = Inventory('laptops.csv')                 

# Verifica promoções
print(inventario.check_promotion_dollars_fast(1339))  
print(inventario.check_promotion_dollars_fast(335))   

# Encontra laptops na faixa de preço de $500 a $1000 
laptops_in_range = inventory.find_laptops_in_price_range(900, 1000)
print("laptops mais baratos:")
for laptop in laptops_in_range:
    print(laptop)

# Encontra o laptop mais barato com pelo menos 8GB de RAM e 256GB 
cheapest_laptop = inventory.find_cheapest_laptop_with_specifications(8, 256)
if cheapest_laptop:
    print("Os laptops mais baratos com especificações:")
    print(cheapest_laptop)
else:
    print("Não encontrado")
```
